# Gradient Descent Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Data Visualisation

```python
x = np.array([1,2,3,4,5,6])
y = np.array([5,6,14,18,25,35])
plt.scatter(x,y)
```

## Main Algorithm

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.05

    for i in range (iterations):
        y_predicted = m_curr * x + b_curr                                                   # y = mx + b
        cost = (1/n) * sum(val**2 for val in (y - y_predicted))                             # average mean square error  
        md = -(2/n)*sum(x*(y-y_predicted))                                                  # partial differenciation of avg mean square error w.r.t m
        bd = -(2/n)*sum(y-y_predicted)                                                      # partial differenciation of avg mean square error w.r.t b
        #updating m_curr and b_curr for each iterations...
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print("m {}, b {}, cost {}, iterations {}".format(m_curr,b_curr,cost,i))



gradient_descent(x,y)



```
