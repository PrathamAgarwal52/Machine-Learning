```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
sns.set(color_codes = True, style = 'whitegrid')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips , hue = 'smoker', kind = 'scatter', size = 'size' , row= 'day' , style='time')
```

```python
sns.scatterplot(x = 'tip' , y = 'total_bill' , data = tips, hue = 'smoker', style = 'time', size = 'size')
```

```python
sns.lineplot(x = 'tip' , y = 'total_bill' , data = tips, hue = 'smoker', style = 'time', size = 'size', dashes = False, markers = False)
```

```python
sns.rugplot(data = tips, x='tip', y = 'total_bill', height = 0.2, color = 'green')
```

```python
sns.kdeplot(data = tips, x= 'tip', y = 'total_bill', hue='smoker', shade = True, bw_adjust = 100, bw_method = 'silverman', cummulative= True, kernaL = 'gau', gridsize = 100, common_norm = True, legend = True)
```

```python
sns.histplot(data=tips, x='total_bill', hue='time', element='step', bins = 30)
```

```python
sns.histplot(data=tips, x='total_bill', hue='time', multiple='stack')
```

```python
sns.stripplot(data = tips, x = 'total_bill', y = 'sex', jitter = True, dodge = True, palette = 'Set2', size = 5, hue = 'smoker')
```
