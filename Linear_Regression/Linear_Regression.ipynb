## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Visualization

```python
x = np.array([1,3,7,9,11,15])
y = np.array([2,6,14,18,28,36])

plt.scatter(x,y)
plt.xlabel("Values of x")
plt.ylabel("Values of y")
plt.show()
x,y
```

## Finding Mean for both X and Y values

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
print("mean of x values:",x_mean,"mean of y values:",y_mean)
```

## Calculating Values of Coefficient(m) and Intercept(b)

```python
num = 0
den = 0
for i in range(len(x)):
    num += (x[i] - x_mean)*(y[i]-y_mean)
    den += (x[i] - x_mean)**2
m = num/den
b = y_mean - (m * x_mean)
print("Intercept is ",float(b),"Coeffiecient is ",float(m))
```

## Predicting the value of y as it is our target variable

```python
y_pred = m * x + b
y_pred
```

## Graphical Representation of results

```python
plt.scatter(x,y,c="orange")
plt.plot(x,y_pred)
plt.scatter(x,y_pred)
```

As we know there are multiple line passing through these points but we want a line which gives minimum error so by using linear regression we got it by putting values of x in this graph we can easily predict values of y
