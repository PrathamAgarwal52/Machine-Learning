```python
import numpy as np
import pandas as pd
```

```python
headers_data = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

df = pd.read_csv(r'C:\Users\lenovo\Desktop\Machine Learning\Data Preprocessing\dataset_1.data', names = headers_data)
df.head(15)
```

```python
df.info()
```

```python
df.replace("?", np.nan, inplace = True)
df.head(10)
```

```python
missing_value = df.isnull()
missing_value
```

```python
for column in headers_data:
    print(column)
    print(missing_value[column].value_counts())
```

```python
df.dtypes
```

```python
mean_nl = df['normalized-losses'].astype(float).mean()
mean_nl
```

```python
df['normalized-losses'].replace(np.nan, mean_nl, inplace = True)
df.head()
```

```python
mean_bore = df['bore'].astype(float).mean()

mean_stroke = df['stroke'].astype(float).mean()

mean_pr = df['peak-rpm'].astype(float).mean()

mean_hp = df['horsepower'].astype(float).mean()

print(mean_bore, mean_stroke, mean_pr, mean_hp)
```

```python
df['bore'].replace(np.nan, mean_bore, inplace = True)
df['stroke'].replace(np.nan, mean_stroke, inplace = True)
df['peak-rpm'].replace(np.nan, mean_pr, inplace = True)
df['horsepower'].replace(np.nan, mean_hp, inplace = True)
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'].replace(np.nan, 'four', inplace = True)
```

```python
missing_value2 = df.isnull()
for column in headers_data:
    print(missing_value2[column].value_counts())
```

```python
df.dropna(subset = 'price', inplace = True)
df.reset_index(drop = True, inplace = True)
```

```python
df.head()
```

```python
final_missing_values = df.isnull()
for column in headers_data:
    print(final_missing_values[column].value_counts())
```
